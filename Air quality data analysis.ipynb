### Session 2. Feature Engineering

````python
# select_dtypes 함수
num_cols = df1.select_dtypes(include=np.number).columns.tolist()
obj_cols = df1.select_dtypes(include=object).columns.tolist()

# target 변수 제거
num_cols.remove('RH')

# 표준화 작업(0 ~ 1)
# 값의 범위가 너무 넓다
scaler = StandardScaler()
df1_obj = df1[obj_cols]
df1_scaled = scaler.fit_transform(df1[num_cols])
df_1_scaled = pd.DataFrame(df1_scaled, index=df1.index, columns=num_cols)

# 데이터프레임 병합
df1 = pd.concat([df_1_scaled, df1['RH'], df1_obj], axis=1)
````

````python
df1.describe()
````

````python
df1 = df1[col]
df1
````

````python
# 월 별 변화량을 보고 싶기 때문에 MONTH 변수 생성
df1['DATE'] = pd.to_datetime(df1.DATE, format='%d-%m-%Y')
df1['MONTH'] = df1['DATE'].dt.month
df1.head(10)
````

````python
# 시간 HOUR 변수 생성
# apply함수 사용
df1['HOUR'] = df1['TIME'].apply(lambda x: int(x.split(':')[0]))
````

### Session 3. Modeling

#### Linear Regression

````python
# Import Packages
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
````

````python
df1
````

````python
X = df1[col1].drop('RH', axis=1) # 14
y = df1['RH'] # 1
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X.head()
````

````python
X_scaled
````

````python
y.head()
````

````python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.4,
                                                    random_state=3)
````

````python
# Data Size
print(X_train.shape)
print(X_test.shape)
````

````python
# Modeling1
model1 = LinearRegression()
model1.fit(X_train, y_train)

print('Intercept:', model1.intercept_)

# 예측값
pred1 = model1.predict(X_test)
pred1
````

````python
y_test
````

````python
# 각 features의 계수값
list(zip(X.columns, model1.coef_))
````

````python
# RMSE
from sklearn import metrics
rmse1 = np.sqrt(mean_squared_error(y_test, pred1))
print(rmse1)
````
